```python
import math
from matplotlib import pyplot as plt
import numpy as np
```

```python
def sigmoid(z):
    return 1./(1.+np.exp(-z))
```

```python
fig,ax=plt.subplots()
```

```python
z=np.linspace(-100,100)
```

```python
fz=sigmoid(z)
```

```python
plt.plot(z,fz)
```

```python
z=np.linspace(-10,10)
fz=sigmoid(z)
plt.plot(z,fz)
```

```python
fig,ax=plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(10)
ax.set_xlabel('z')
ax.set_ylabel('f(z)')
ax.grid(True,which='major')
ax.plot(z,fz)
ax.legend(['f(z)=1/(1+exp(-z))'])
```
